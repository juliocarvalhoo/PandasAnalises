```python
"""UTILIZANDO RELPLOT PARA REPRESENTAR UM GRAFICO DE LINHAS"""
"""MEDIA ESTATISTICA"""
sns.relplot(pinguim_info_df, y = "Flipper Length (mm)", x ="Body Mass (g)", kind= "line")
```

```python
"""Regresão Linear"""
sns.lmplot(pinguim_info_df, x = "Body Mass (g)", y = "Flipper Length (mm)", line_kws={'color': 'green'} )
```

```python
sns.lmplot(pinguim_info_df, x = "Body Mass (g)", y = "Flipper Length (mm)", hue = "Sex" )
```

```python
"""DISPLOT PARA REPRESENTAR HISTOGRAMAS, NORMALMENTE USAMOS UMA INFORMACAO PARA AGRUPAR"""
sns.displot(pinguim_info_df, x ="Body Mass (g)", kde=True)
plt.show()
```

```python
pinguim_info_df["Species"].unique()
```

```python
"""UTILIZANDO CATPLOT PARA REPRESENTAR DADOS CATEGORICOS"""
sns.catplot(data = pinguim_info_df, x = "Species", y = "Culmen Length (mm)", height = 6)
plt.gca().set_xticks([0,1,2],["Adelie","Chinstrap","Gentoo"])
plt.title("Tamanho do bico x Espécie")
plt.xlabel("Espécie")
plt.ylabel("Tamanho do bico")
plt.show()
```

```python
sns.catplot(data = pinguim_info_df, x = "Species", y = "Culmen Length (mm)", hue = "Sex", height = 5)
plt.gca().set_xticks([0,1,2],["Adelie","Chinstrap","Gentoo"])
plt.title("Tamanho do bico x sexo da espécie")
plt.xlabel("Espécie")
plt.ylabel("Tamanho do bico")
plt.show()
```

```python
"""" UTILIZANDO O KIND = "VIOLIN" PARA SEPARAR AS INFORMAÇÕES"""
sns.catplot(data = pinguim_info_df, kind = "violin", x = "Species", y = "Culmen Length (mm)", hue = "Sex", height = 5)
plt.gca().set_xticks([0,1,2],["Adelie","Chinstrap","Gentoo"])
plt.title("Tamanho do bico x sexo da espécie")
plt.xlabel("Espécie")
plt.ylabel("Tamanho do bico")
plt.show()
```

```python
"""" GRAFICO DE BARRAS USANDO KIND = "BAR" INTERVALO DE CONFIANÇA, POR PADRAO ELE PEGA A MÉDIA """
sns.catplot(data = pinguim_info_df, kind = "bar", x = "Species", y = "Culmen Length (mm)", hue = "Sex", height = 5)
plt.gca().set_xticks([0,1,2],["Adelie","Chinstrap","Gentoo"])
plt.title("Tamanho do bico x sexo da espécie")
plt.xlabel("Espécie")
plt.ylabel("Tamanho do bico")
plt.show()
```

```python
"""Aplicando o JOINTPLOT PRA JUNTAR DUAS INFORMAÇÕES"""

sns.jointplot(data = pinguim_info_df, x = "Flipper Length (mm)", y = "Culmen Length (mm)", hue = "Species")
plt.ylabel("Altura bico")
plt.xlabel("Nadadeira")
plt.show()
```

```python
"""TRANFORMANDO O JOINTPLOT EM REGRESSÃO LINEAR COM HISTOGRAMAS"""
sns.jointplot(data = pinguim_info_df, x = "Flipper Length (mm)", y = "Culmen Length (mm)", kind = "reg")
plt.ylabel("Altura bico")
plt.xlabel("Nadadeira")
plt.show()
```

```python
sns.lmplot(data = pinguim_info_df, x = "Flipper Length (mm)", y = "Culmen Length (mm)", hue = "Species")
plt.ylabel("Altura bico")
plt.xlabel("Nadadeira")
plt.show()
```

```python
"""PAIRPLOT - ANALIZA TODAS AS INFORMAÇÕES DO DATAFRAME, ÓTIMO PARA ANALISAR OS DADOS"""

sns.pairplot(data = pinguim_df, hue = "species")
```

```python
g = sns.PairGrid(data = pinguim_df, hue = "species", corner = True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```python
""" SEABORN USANDO A MASSA PARA COLORAÇÃO GRADIENTE DESTACAR O PESO"""

sns.relplot(pinguim_df, x = "body_mass_g", y = "culmen_depth_mm", hue = "body_mass_g")
plt.show()
```
