```python
"""Treinamento de Analise Com PANDAS"""
import pandas as pd
from datetime import date

pathCustomer = "./customer_shopping_data.csv"

with open(pathCustomer,"r",encoding = "utf-8") as file:
    customer_df = pd.read_csv(file)
```

```python
#MEDIA DE COMPRAS POR SHOPPING  #RESETAR INDEX
customer_df.groupby(["shopping_mall"])[["price"]].mean().reset_index().rename(columns={"shopping_mall": "Shopping","price":"Preço"})
```

```python
#LINHAS ALEATÓRIAS
customer_df.sample(5)
```

```python
#PRIMEIRO QUARTIL DE 22
#CONVERTER DATA
customer_df["invoice_date"] = pd.to_datetime(customer_df["invoice_date"], format="%d/%m/%Y")
# customer_df["invoice_date"] = customer_df["invoice_date"].dt.strftime("%d/%m/%Y")
```

```python
# TODOS OS DADOS DE 22
# MANEIRA MAIS FACIL #DT.YEAR == 2022
year_df = customer_df[(customer_df["invoice_date"] >="2022-01-01") & (customer_df["invoice_date"] <= "2022-12-31")]
# year_df["invoice_date"] = (year_df["invoice_date"]).dt.strftime("%d/%m/%Y")
year_df.head()
```

```python
#CRIANDO TABELA QUANTIS
year_df["quantis"] = pd.qcut(year_df["invoice_date"], q=4,labels = False)
```

```python
#SOMATORIA DO PREÇO POR QUANTIS / VISITAS / MEDIA DE PREÇO POR VISITAS
shoppings_df = year_df.groupby(["shopping_mall","quantis"])[["price","quantity"]].agg({"price":"sum","quantity":"count"}).reset_index()
shoppings_df["mean_for_price"]= (shoppings_df["price"]/shoppings_df["quantity"]).round(2)
shoppings_df["price"] = shoppings_df["price"].round()
shoppings_df.rename(columns = {"shopping_mall":"Shopping","quantis":"Quantis","price":"Preço","quantity":"Compra","mean_for_price":"Media/Preço"})
```
